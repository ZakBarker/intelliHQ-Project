```python
def remove_noise(rr, before=3, after=3):
    results = []
    for i in range(len(rr)):
        b = []
        a = []
        if i < before and i > 0:
            b = rr[0:i]
        elif i == 0:
            b = []
        else:
            b = rr[i-before: i]
            
        if i >= len(rr) - after:
            a = rr[i+1:]
        else:
            a = rr[i+1: i+1+after]
        a.extend(b)
        mean = np.mean(a)
        std = np.mean(a)
        z = (rr[i] - mean) / std
        if abs(z) < 0.5:
            results.append(rr[i])
    return results
         
```

```python
import numpy as np

def process_group(group, threshold=2.5):
    """
    Finds outliers in group and replaces it
    """
    mean = np.mean(group)
    std = np.std(group)
    for i in range(len(group)):
        z = abs((group[i] - mean) / std)
        if z >= threshold:
            group[i] = mean
    return group
            
        

def group_peaks(peaks, min_bpm=50, max_bpm=115):
    groups = []
    current_group = []
    for i in range(len(peaks)-1):
        p1 = peaks[i]
        p2 = peaks[i+1]
        rr = p2 - p1
        bpm = rr / 300 * 60
        if min_bpm < bpm and bpm < max_bpm:
            current_group.append(rr)
        else:
            if current_group is not None:
                if len(current_group) >= 3:
                    groups.append(remove_noise(current_group))
                current_group = []
    if current_group is not None:
        groups.append(remove_noise(current_group))
    return groups

def time_bucket(r_peaks, hours=1):
    # 300hz * 60s * 60m / hours
    ticks = 300 * 60 * 60 * hours
    groups = []
    curr_group = []
    curr_hour = 1
    for x in r_peaks:
        x = x - r_peaks[0]
        if x > curr_hour * ticks:
            groups.append(curr_group)
            curr_group = []
            curr_group.append(x)
            curr_hour += 1
        else:
            curr_group.append(x)
    groups.append(curr_group)
    return groups

def get_rmssd_hrv(rr, sections=3):
        hrv_total = []
        hrv_sections = []
        for j in range(len(rr) // sections):
            values = rr[j*sections:j*sections+sections]
            rr_delta = []
            for i in range(len(values)-1):
                d = (values[i] - values[i+1]) ** 2
                rr_delta.append(d)
            hrv = math.sqrt(1 / (len(values)-2) * sum(rr_delta))
            hrv_sections.append(hrv)

        for i in range(len(rr)-1):
            d = (rr[i] - rr[i+1]) ** 2
            hrv_total.append(d)
        hrv = math.sqrt(1 / (len(rr)-2) * sum(hrv_total))
        return hrv, hrv_sections
```

```python
from scipy import signal
def plot_lombscargle(sig, low=0.02, high=0.08):
    nout = 1000
    f = np.linspace(low, high, nout)
    x = range(0,len(sig))
    psd = signal.lombscargle(x, sig, f, normalize=False)
    plt.figure(figsize=(10, 4))
    plt.plot(f,psd)
    plt.title('Power Spectral Density (PSD) - Lomb-Scargle')
    plt.xlabel('Frequency')
    plt.ylabel('Power')
    plt.tight_layout()
```

```python
def create_rr_buckets(buckets):
    g = [None] * len(buckets)
    flatten = lambda l: [item for sublist in l for item in sublist]
    for bucket in buckets:
        for i in range(len(bucket)):
            p_groups = group_peaks(bucket[i])
            p_groups = flatten(p_groups)
            if len(p_groups) > 2:
                if g[i] is None:
                    g[i] = [p_groups]
                else:
                    g[i].append(p_groups)
    return g
```

```python
times = ["{}min".format(15 * (i+1)) for i in range(len(buckets))]
```

```python
def Calculate_Features(R_peaks, fs=1, decim=2):
    R_peaks = np.array(R_peaks)
    # Calculating SDNN
    R_peaks2 = R_peaks / fs  # Turn R-Peak locations to time stamps
    R_td = np.diff(R_peaks2)
    MeanRR = np.round(np.mean(R_td) * 1e3, decim)
    SDNN = np.round(np.std(R_td) * 1e3, decim)

    # Calculating SDANN
    timejump = 300  # 5 minutes
    timestamp = timejump
    runs = int(R_peaks2[-1] / timestamp)
    SDNN_5 = np.zeros(runs)
    i = 0
    while (timestamp <= timejump * runs):
        section = R_peaks2[R_peaks2 <= timestamp]
        R_peaks2 = R_peaks2[R_peaks2 > timestamp]
        timestamp += timejump
        R_td_5 = np.diff(section)
        SDNN_5[i] = np.std(R_td_5)
        i += 1

    SDANN = np.round(np.mean(SDNN_5) * 1e3, decim)

    # Calculating pNN50                      pNN50 = (NN50 count) / (total NN count)
    total_NN = len(R_peaks)
    NN_50 = np.diff(R_td)
    count = 0
    for i in range(len(NN_50)):
        if NN_50[i] > 0.050:
            count = count + 1

    pNN50 = np.round((count / total_NN * 100), decim)

    # Calculating RMSSD

    R_peaks3 = R_peaks / fs * 1000  # Turn R-Peak locations to time stamps
    RMSSD = np.round((np.sqrt(np.sum(np.power(np.diff(R_peaks3), 2)) / (len(R_peaks3) - 1))), decim)

    return (SDNN, SDANN, MeanRR, RMSSD, pNN50)
```
